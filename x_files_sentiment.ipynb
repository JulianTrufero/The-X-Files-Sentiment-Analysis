```python
import pandas as pd
import re
```

```python
s = x_file.to_dict()
```

```python
s2 = s.values()
```

```python
script = list(list(s2)[0].values())
```

```python
script
```

```python
scene = []
for i in script:
    match = re.findall("\SCENE\s\d.*", i)
    if match:
        scene.append(i)
```

```python
scene
```

```python
def scullyphrases():
    scully = []
    for s in script:
        match = re.findall("\SCULLY\S.*", s)
        if match:
            scully.append(s)
    scully_dic = {'Scully': [i[8:] for i in scully]}
    scully_phrases = pd.DataFrame(scully_dic)
    return scully_phrases
```

```python
scullyphrases()
```

```python
def mulderphrases():
    mulder = []
    for s in script:
        match = re.findall("\SULDER\S.*", s)
        if match:
            mulder.append(s)
    mulder_dic = {'Mulder': [i[8:] for i in mulder]}
    mulder_phrases = pd.DataFrame(mulder_dic)
    return mulder_phrases
```
